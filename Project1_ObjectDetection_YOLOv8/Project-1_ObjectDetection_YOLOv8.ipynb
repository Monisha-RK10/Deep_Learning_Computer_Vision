```python
#this code implements object detection via YOLO model on a custom dataset 
#(https://github.com/entbappy/YOLO-v8-Object-Detection/blob/main/data.zip). 
#It involves the following steps:
#data folders (train, validation, and test) for images
#data annotation (create labels for train and validation via 'labelImg', type 'labelImg' in Anaconda prompt)
#data structuring (keep all the labels in a 'labels' folder for train and validation images)
#train model
#test model

#import libraries and YOLO model using internal python package for YOLO
from ultralytics import YOLO
#select the model YOLOv8 from YOLO
model = YOLO("yolov8n.pt")
```

```python
#load the file 'yaml' that contains the path of train and validation images, information of number (5) and names (Helmet, 
#Goggles, Jacket, Gloves, Footwear) of classes
data_yaml_path = r'D:\ComputerVision_Projects\ObjectDetection_YOLOv8\data.yaml'
#set current directory to 'ObjectDetection_YOLOv8' folder
%cd D:\ComputerVision_Projects\ObjectDetection_YOLOv8
#train the model on train and evaluate on the validation dataset
results = model.train(data= data_yaml_path, epochs=30)
```

```python
#set the path where the trained model saved the 'best' weights 
%cd D:\ComputerVision_Projects\ObjectDetection_YOLOv8\runs\detect\train\weights
#pass the 'best' weights to the YOLO model
model_predict = YOLO("best.pt")
```

```python
#load the path of the test image
img_path = r'D:\ComputerVision_Projects\ObjectDetection_YOLOv8\data\test\images'
save_path =r'D:\ComputerVision_Projects\ObjectDetection_YOLOv8\data\test\labels'
#perform the inference for test data on the best weights
model_predict.predict(img_path, save = True, project=save_path)
```
